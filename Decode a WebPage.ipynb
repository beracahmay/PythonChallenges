#### Challenge

Use the BeautifulSoup and requests Python packages to print out a list of all the article titles on the New York Times homepage.

```python
#Trial 1

# Import the relevant libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")


# Scraping 
url="https://www.nytimes.com/"
nyt_data = requests.get(url)

soup = BeautifulSoup(nyt_data.text,"html.parser")
nyt_data=soup.findAll('section',attrs = {'class': 'story-wrapper css-g00kst'})
nyt_titles= soup.find_all("section", {"class" : lambda L: L and L.startswith('story-wrapper')})

titles=[]

for title in nyt_titles:
    title=title.find("h3", {"class" : lambda L: L and L.startswith('indicate-hover')}).text
    titles.append(title)
    
titles
```

```python
# Refining the initial Trial

url="https://www.nytimes.com/"
nyt_data = requests.get(url)

titles2=[]
titles3=[]

soup = BeautifulSoup(nyt_data.text,"html.parser")
#nyt_data=soup.findAll('section',attrs = {'class': 'story-wrapper css-g00kst'})
nyt_titles2= soup.find_all("div", {"class": "css-17tuifc e1ppw5w20"})


for l in nyt_titles2[:11]:
    for li in l.find_all("section", {"class" : lambda L: L and L.startswith('story-wrapper')}):
        title=li.find("h3", {"class" : lambda L: L and L.startswith('indicate-hover')}).text
        titles2.append(title)
        

# Removing Duplixates
for i in titles2:
    if i not in titles3:
        titles3.append(i)
        
titles3
```
